## Test: Deck Class

### Correct deck size

```python
deck = Deck()
print("Number of cards in new deck:", len(deck.show_cards()))
assert len(deck.show_cards()) == 52, "Deck should contain 52 cards."
```

### Drawing cards reduces deck size

```python
deck = Deck()
hand = deck.draw(5)
remaining = len(deck.show_cards())

print("Drawn cards:", hand)
print("Remaining cards:", remaining)

assert len(hand) == 5, "Should draw exactly 5 cards."
assert remaining == 47, "Deck should have 47 cards left."
```

### No duplicate cards

```python
deck = Deck()
drawn = deck.draw(52)
unique_cards = set(drawn)

print("Unique cards drawn:", len(unique_cards))
assert len(unique_cards) == 52, "All cards should be unique."
```

### Deck is empty after drawing all cards

```python
deck = Deck()
deck.draw(52)
remaining = len(deck.show_cards())

print("Remaining cards after full draw:", remaining)
assert remaining == 0, "Deck should be empty after drawing all cards."
```

### Multiple decks are shuffled differently

```python
deck1 = Deck()
deck2 = Deck()

order1 = [str(c) for c in deck1.show_cards()]
order2 = [str(c) for c in deck2.show_cards()]

same_order = order1 == order2
print("Decks have identical order:", same_order)
assert not same_order, "Decks should be shuffled differently."
```

## Test: Player Class


### Basic initialization

```python
player = Player(name="Bot1", stack=1000, position=0)

print("Name:", player.name)
print("Stack:", player.stack)
print("Position:", player.position)
print("Initial state -> Folded:", player.folded, "| All-in:", player.all_in)

assert player.name == "Bot1"
assert player.stack == 1000
assert player.position == 0
assert player.folded is False
assert player.all_in is False
assert player.current_bet == 0
assert player.total_bet == 0
```

### Placing a normal bet

```python
player = Player("Bot2", stack=500)
bet_amount = player.bet(100)

print("Bet amount:", bet_amount)
print("Stack after bet:", player.stack)
print("Current bet:", player.current_bet)
print("Total bet:", player.total_bet)

assert bet_amount == 100
assert player.stack == 400
assert player.current_bet == 100
assert player.total_bet == 100
assert not player.all_in
```

### All-in bet (bet > stack)

```python
player = Player("Bot3", stack=50)
bet_amount = player.bet(100)

print("Bet amount (should equal stack):", bet_amount)
print("Stack after bet:", player.stack)
print("All-in status:", player.all_in)

assert bet_amount == 50
assert player.stack == 0
assert player.all_in
```

### Reset between hands

```python
player = Player("Bot4", stack=800)
player.bet(100)
player.folded = True
player.hand = [Card('A', 'Spades'), Card('K', 'Hearts')]
player.reset_for_new_hand()

print("After reset -> Folded:", player.folded)
print("After reset -> Stack:", player.stack)
print("After reset -> Hand:", player.hand)
print("After reset -> Bets:", player.current_bet, player.total_bet)

assert player.folded is False
assert player.current_bet == 0
assert player.total_bet == 0
assert player.hand == []
```

### Is_active logic

```python
active_player = Player("Active", stack=200)
folded_player = Player("Folded", stack=200)
folded_player.folded = True
allin_player = Player("AllIn", stack=0)
allin_player.all_in = True

print("Active:", active_player.is_active())
print("Folded:", folded_player.is_active())
print("All-in:", allin_player.is_active())

assert active_player.is_active()
assert not folded_player.is_active()
assert not allin_player.is_active()
```
